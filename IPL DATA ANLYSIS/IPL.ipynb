```python
 #Finding the team won by batting first
batting_first = df[df['win_by_runs']!=0]
batting_first.head()
```

```python
#making an histrogram graph
plt.figure(figsize=(10,6))
plt.hist(batting_first['win_by_runs'],color='b')
plt.title('Distribution of runs in batting first')
plt.xlabel('Total runs')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(batting_first['win_by_runs'],bins =range(0,max(batting_first['win_by_runs'])+5,5),color='r',edgecolor='black') #bins=range(0, max(batting_first['win_by_runs']) + 5, 5) specifies the bins for the histogram. It creates bins with a width of 5 runs, starting from 0 and ending at the maximum value in the 'win_by_runs' column plus 5. This ensures that the data is divided into bins of 5-run intervals.
plt.title('Distribution of runs in batting first')
plt.xlabel('Total runs')
plt.show()
```

```python
#no of winners while batting first
batting_first['winner'].value_counts()
```

```python
top_winners = batting_first['winner'].value_counts().head(3)
winner_names = top_winners.index
winner_counts = top_winners.values
```

```python
#bar plotting the top 3 no of teams with most wins in batting first
plt.figure(figsize=(10,6))
plt.bar(winner_names,winner_counts,color=['blue','orange','red'])
plt.title('Top 3 teams with most wins in batting first')
plt.xlabel('Teams')
plt.ylabel('No of wins')
plt.show()
```

```python
#making an bar chart 
plt.figure(figsize=(10,5))
plt.pie(winner_counts,labels=winner_names,autopct="%0.1f%%")
plt.title('Top 3 teams with most no of wins while batting first')
plt.show()
```
