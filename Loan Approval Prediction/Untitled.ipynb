### dealing with null values  'Categorial values'


```python
df.drop(['Loan_ID','Dependents'],axis=1,inplace=True)
```

```python
df
```

```python
cols = df[['Gender','Married','Self_Employed']]
for i in cols:
    df[i].fillna(df[i].mode().iloc[0],inplace=True)
```

```python
df.isnull().sum()
```

### dealing with null values  'Numerical values'


```python
n_cols = df[['LoanAmount','Loan_Amount_Term','Credit_History']]
for i in n_cols:
    df[i].fillna(df[i].mean(axis=0),inplace=True)
```

```python
df.isnull().sum()
```

### Visualization

```python
def bar_chart(col):
    Approved = df[df['Loan_Status']=='Y'][col].value_counts()
    Disapproved = df[df['Loan_Status']=='N'][col].value_counts()
    df1 = pd.DataFrame({'Approved': Approved, 'Disapproved': Disapproved})
    df1.plot(kind='bar')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.title('Loan Approval Status by ' + col)
    plt.show()
```
